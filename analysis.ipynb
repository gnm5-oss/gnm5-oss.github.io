```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

df = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/ufo-scrubbed-geocoded-time-standardized-00.csv")

df.columns = [
    "datetime","city","state","country","shape",
    "duration_seconds","duration_hours_min","comments",
    "date_posted","latitude","longitude"
]

df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")
df = df[df["country"] == "us"]
df = df.dropna(subset=["datetime"])

df["year"] = df["datetime"].dt.year

# Plot 1
state_counts = df.groupby("state").size().reset_index(name="count")

chart1 = alt.Chart(state_counts.sort_values("count", ascending=False).head(15)).mark_bar().encode(
    x="count:Q",
    y=alt.Y("state:N", sort="-x"),
    color=alt.Color("count:Q", scale=alt.Scale(scheme="blues"))
)

# Plot 2 (line)
shape_counts = df.groupby(["year","shape"]).size().reset_index(name="count")

chart2 = alt.Chart(shape_counts).mark_line().encode(
    x="year:Q",
    y="count:Q",
    color="shape:N"
)

# SAVE FILES
chart1.save("ufo_plot1.html")
chart2.save("ufo_plot2.html")
```
